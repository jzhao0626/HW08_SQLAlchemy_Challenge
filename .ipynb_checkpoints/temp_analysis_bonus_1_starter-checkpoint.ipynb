# Bonus: Temperature Analysis I

```python
import pandas as pd
from datetime import datetime as dt
```

```python
# "tobs" is "temperature observations"
df = pd.read_csv('Resources//hawaii_measurements.csv')
df
```

```python
# Convert the date column format from string to datetime
df['date'] = pd.to_datetime(df['date'])
df
```

```python
# Set the date column as the DataFrame index
sorted_df = df.set_index('date', drop=False)
sorted_df
```

```python
# Drop the date column
droped_df = sorted_df.drop('date', 1)
droped_df
```

### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Identify the average temperature for June
jun_df = droped_df.filter(like = '-06-', axis=0).groupby('date').mean()
jun_df['tobs'] = round(jun_df['tobs'], 1)
jun_df
```

```python
# Identify the average temperature for December
dec_df = droped_df.filter(like = '-12-', axis=0).groupby('date').mean()
dec_df['tobs'] = round(dec_df['tobs'], 1)
dec_df
```

```python
# Create collections of temperature data
jun = jun_df['tobs']
dec = dec_df['tobs']
```

```python
# Run unpaired t-test
stats.ttest_ind(jun, dec)
```

### Analysis

It would be more reasonable to use an unpaired t-test because the temperature of June and the temperature of December would be considered as independent or unrelated to each other. From the t-score we got from the statistic much greater than 3 or 4(which is considered as statistically different), we can see that there is a huge difference among the two groups of samples, and the p-value much lower than 0.05 confirmed that this result is not happening by chance; in other word, it is not likely to be a random result. So as a result, there is a statistically significant difference between the temperature in June and the temperature in December.
